# Imports

```python
import numpy as np
import pandas as pd
```

# Load Data

```python
data = pd.read_csv( './daylio_export.csv' )
data
```

```python
day = ' ' + data['date'] + ' '
year_day = data['year'].values.astype( str ) + day
date = np.core.defchararray.add( year_day.values.astype( str ), data['time'].values.astype( str ) )
```

```python
datetime = pd.to_datetime( date )
time = datetime.astype( np.int64 ) / 1e9 / 3600
```
