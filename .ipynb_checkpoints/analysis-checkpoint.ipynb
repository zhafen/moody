# Imports

```python
import datetime
import dateutil
import numpy as np
import pandas as pd
import time
```

```python
import verdict
```

```python
from mpl_toolkits.axes_grid1 import make_axes_locatable
import matplotlib.dates as mdates
```

# Load and Prep Data

```python
data = pd.read_csv( './daylio_export.csv' )
```

```python
data['timestamp'] = pd.to_datetime( data['full_date'] + ' ' + data['time'] )
```

```python
mood_mapping = {
    'Angry': -2,
    'anxious': -1.25,
    'awful': -2,
    'bad': -1,
    'beyond anxious': -2,
    'content': 1,
    'distracted': -0.75,
    'exhausted': -1.75,
    'frustrated': -1.25,
    'good': 1,
    'in pain': -1.75,
    'meh': -0.5,
    'okay': 0.,
    'rad': 2.,
    'tired': -1,
}

data['mood_int'] = [ mood_mapping[mood] for mood in data['mood'] ]
```

```python
# Process data by making individual data frames

activities_data = {}

for i in data.index:
    
    # Get the activities out
    activities = data['activities'][i].split( ' | ')

    for activity in activities:
        try:
            activities_data[activity].append( data.loc[i] )
        except KeyError:
            activities_data[activity] = [ data.loc[i] ]
```

```python
# Turn into pandas DataFrames
for key, item in activities_data.items():
    activities_data[key] = pd.concat( activities_data[key], axis=1 ).transpose()
```

```python
# Turn into a verdict Dict
activities_data = verdict.Dict( activities_data )
```

# Mood Overview

### Get bins for data

```python
# Unique mood values
unique_mood_values = np.unique( list( mood_mapping.values() ) )

unique_mood_values
```

```python
# Mood bins

delta_mood = unique_mood_values[1:] - unique_mood_values[:-1]

mood_bins = unique_mood_values[:-1] + delta_mood / 2.
mood_bins = np.insert( mood_bins, 0, unique_mood_values[0] - delta_mood[0]/2. )
mood_bins = np.append( mood_bins, unique_mood_values[-1] + delta_mood[-1]/2. )

mood_bins
```

```python
# Override and use standard mood bins

mood_bins = np.arange( -2.5, 2.51, 1. ) + 0.01
```

```python
mpl_dates = mdates.epoch2num( data['timestamp'].astype( int ) // 1e9 )

mpl_date_bins = []
n = 0
below_end = True
while below_end:
    
    # Extract the bin itself
    date = data['timestamp'].dt.date.values[-1] + dateutil.relativedelta.relativedelta(months = n) - dateutil.relativedelta.relativedelta( seconds = 1. )
    
    # Format
    num = mdates.date2num( date )
    
    # Store
    mpl_date_bins.append( num )
    
    # Check if done
    below_end = num < mpl_dates.max()
    
    n += 1
    
mpl_date_bins = np.array( mpl_date_bins )
```

```python
def plot_data_vs_time(
    data,
    mpl_date_bins,
    key = 'mood_int',
    ax = None,
    x_tick_locator=mdates.MonthLocator(),
    x_tick_formatter=mdates.DateFormatter('%y/%m/%d'),
    **kwargs
):
    
    if ax is None:
        fig = plt.figure( figsize=(25,11), facecolor='white' )
        ax = plt.gca()

    try:
        x_data = mdates.epoch2num( data['timestamp'].astype(int) // 1e9 )
    except TypeError:
        x_data = mdates.epoch2num( [ time.mktime( _.timetuple() ) for _ in data['timestamp'].values ] )

    hist2d, x_edges, y_edges = np.histogram2d( x_data, data[key].astype( float ), bins=[ mpl_date_bins, mood_bins ] )

    hist_x, x_edges = np.histogram( x_data, x_edges )

    # Make the y-axis conditional (so we don't depend on how frequently daylio was answered)
    hist2d /= hist_x[:,np.newaxis]

    im = ax.pcolormesh(                                                         
        x_edges,                                                                
        y_edges,                                                                
        hist2d.transpose(),                                                     
        **kwargs
    ) 

    # Date formatting
    ax.xaxis.set_major_locator(x_tick_locator)
    ax.xaxis.set_major_formatter(x_tick_formatter)
    ax.yaxis.set_tick_params( labelsize=24 )
    ax.xaxis.set_tick_params( labelsize=24 )

    # Colorbar
    divider = make_axes_locatable(ax)
    cax = divider.append_axes("right", pad=0.05, size='5%')
    cbar = plt.colorbar(im, cax=cax, )

    # Labels
    ax.set_title( 'Fraction of entries in a given mood', fontsize=26 )
    ax.set_xlabel( 'Date', fontsize=26 )
    ax.set_ylabel( 'Mood', fontsize=26 )
```

```python
plot_data_vs_time(
    data = data,
    mpl_date_bins = mpl_date_bins,
)
```

```python
def get_mpl_date_bins(
    date_range,
    **kwargs
):

    custom_mpl_date_bins = []
    n = 0
    below_end = True
    while below_end:

        f_date_range = pd.to_datetime( pd.Series( date_range ) ).dt.date
        
        used_kwargs = {}
        for key, item in kwargs.items():
            used_kwargs[key] = item * n

        # Extract the bin itself
        date = f_date_range[0] + dateutil.relativedelta.relativedelta(**used_kwargs) - dateutil.relativedelta.relativedelta( seconds = 1. )
        
        # Format
        num = mdates.date2num( date )

        # Store
        custom_mpl_date_bins.append( num )

        # Check if done
        below_end = num < mdates.date2num( f_date_range[1] )

        n += 1

    custom_mpl_date_bins = np.array( custom_mpl_date_bins )
    
    return custom_mpl_date_bins
```

```python
fig = plt.figure( figsize=(25,11), facecolor='white' )
ax = plt.gca()

date_range = [ '2019-02-01','2019-02-28' ]

custom_mpl_date_bins = get_mpl_date_bins( date_range, days=1 )

plot_data_vs_time(
    data = data,
    mpl_date_bins = custom_mpl_date_bins,
    ax = ax,
    x_tick_locator = mdates.DayLocator(),
    x_tick_formatter = mdates.DateFormatter('%d')
)

ax.set_title( 'Fraction of entries in a given mood for {} to {}'.format( date_range[0], date_range[1] ), fontsize=26 )
```

# Histograms for given activities

```python
# Mean moods and sorting by them
means = activities_data.inner_item( 'mood_int' ).mean()

sorted_tuples = sorted( list( means.items() ), key=lambda x: x[1] )[::-1]

sorted_as, means = list( zip( *sorted_tuples ) )
```

```python
for key in sorted_as:

    fig = plt.figure( figsize=(15,7), facecolor='white' )
    ax = plt.gca()
    
    item = activities_data[key]

    ax.hist( item['mood_int'], bins=mood_bins, normed=True )

    ax.set_xlim( -2.1, 2.1 )

    ax.set_title( '{}, n={}'.format( key, len( item['mood_int'] ) ), fontsize=28 )
    ax.set_xlabel( 'Mood', fontsize=24 )
    ax.set_ylabel( 'Frequency', fontsize=24 )
```
